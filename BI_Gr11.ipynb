```python
path='/content/drive/MyDrive/Colab Notebooks/Store_Data/StoreDataCleaned.csv'
data=pd.read_csv(path)
```

```python
customers=data['customer_name'].unique().tolist()
```

```python
transactions = []

for customer in tqdm(customers, desc="Processing customers", unit="customer"):
    ordeDF = data[(data['customer_name'] == customer) & ((data['year']==2014)| (data['year']==2013))]
    items = ordeDF['product_name'].unique().tolist()
    transactions.append(items)
```

```python
len(transactions)
```

```python
# Convert transactions to a one-hot encoded DataFrame
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
df.shape
```

```python
from scipy.sparse import csr_matrix
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# Convert the transaction data to a sparse matrix
sparse_df = csr_matrix(df.values)

# Convert the sparse matrix to a DataFrame
dense_df = pd.DataFrame.sparse.from_spmatrix(sparse_df, columns=df.columns)

# Run Apriori algorithm
frequent_itemsets = apriori(dense_df, min_support=0.01, use_colnames=True)

# Filter out less interesting rules early to reduce memory usage
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

# Display the frequent itemsets and rules
print("\nFrequent Itemsets:")
print(frequent_itemsets)

print("\nAssociation Rules:")
print(rules)
```
